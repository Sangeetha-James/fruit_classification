```python
#cols_with_zeros=["shape","color","taste","fruit_name"]
cols_with_zeros=["shape","color","taste"]
```

```python
fruit[cols_with_zeros]=fruit[cols_with_zeros].replace(0,np.nan)
```

```python
#since here  shape is categorical but encoded as numbers, mode is arguably better than median.


fruit['shape'].fillna(fruit['shape'].mode()[0],inplace=True)
```

```python
sns.distplot(fruit['color'])
```

```python
fruit['color'].fillna(fruit['color'].median(),inplace=True)
```

```python
sns.distplot(fruit['taste'])
```

```python
#Mode (best)

#Explicitly chooses the most common taste value

#Respects the discrete nature of the feature

#Avoids inventing “middle” values that don’t really exist
#mode() will return two values here ,but we need the 1st value which is 1,so mode()[0]



fruit['taste'].fillna(fruit['taste'].mode()[0],inplace=True)
```

```python
x=fruit.drop('fruit_name',axis=1)
x
```

```python
y=fruit["fruit_name"]
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
model=RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
train_data_predict=model.predict(x_train)
train_data_accuracy=accuracy_score(train_data_predict,y_train)
print("The accuracy of training data  is",train_data_accuracy)

```

```python
test_data_predict=model.predict(x_test)
test_data_accuracy=accuracy_score(test_data_predict,y_test)
print("The accuracy of testing data  is",test_data_accuracy)
```

```python
cm=confusion_matrix(test_data_predict,y_test)
cm
```

```python
fruit_name=["pomegranate","pear","pineapple","custard apple","papaya","kiwi","blueberry","coconut","banana","lychee","mango","grape","watermelon","dragon fruit","apple","plum","orange","guava","cherry","strawberry"]
```

```python
# Predict (already class labels)
test_data_pred = model.predict(x_test)

# Confusion matrix (true first, pred second)
cm = confusion_matrix(y_test, test_data_pred)

plt.figure(figsize=(16, 14))
sns.heatmap(
    cm,
    cmap="Blues",
    xticklabels=fruit_name,
    yticklabels=fruit_name
)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.xticks(rotation=45, ha="right")
#Rotates the x-axis tick labels 45° and right-aligns them.
#This is super useful when labels are long (dates, names, categories) and would otherwise overlap.
plt.tight_layout()
#Automatically adjusts spacing (margins, padding) so labels, titles, and ticks don’t get cut off.
plt.show()

```
